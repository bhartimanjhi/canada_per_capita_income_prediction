```python
#Predict canada's per capita income in year 2020.

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# First, check the actual column names
print(df.columns)

# Rename the column with special characters
df.columns = [col if col != 'per capita income (US$)' else 'per_capita_income' for col in df.columns]

# Verify the column names changed
print(df.columns)

# Now plot with the renamed column
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df['year'], df['per_capita_income'], color='green', marker='+')

# Alternatively, you can use the original column name directly:
# plt.scatter(df['year'], df['per_capita_income_(US$)'], color='green', marker='+')
```

```python
new_df = df.drop('per_capita_income', axis = 'columns')
new_df
```

```python
per_capita_income=df.per_capita_income
per_capita_income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, per_capita_income)
```

```python
reg.predict([[2020]])
```
